# Define the Model

A GAN is comprised of two adversarial networks, a discriminator and a generator.

## Discriminator

Your first task will be to define the discriminator. This is a convolutional classifier like you've built before, only without any maxpooling layers. To deal with this complex data, it's suggested you use a deep network with **normalization**. You are also allowed to create any helper functions that may be useful.

#### Exercise: Complete the Discriminator class
* The inputs to the discriminator are 32x32x3 tensor images
* The output should be a single value that will indicate whether a given image is real or fake


```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
# convolution with optional batch normalization
def conv(in_channels, out_channels, kernel_size, stride = 2, padding = 1, batch_norm = True):
    layers = []
    conv_layer = nn.Conv2d(in_channels, out_channels, kernel_size,
                           stride = stride, padding = padding, bias = False)
    layers.append(conv_layer)
    
    if batch_norm:
        layers.append(nn.BatchNorm2d(out_channels))
    return nn.Sequential(*layers)
```

```python
# deconvolution with optional batch normalization
def deconv(in_channels, out_channels, kernel_size, stride = 2, padding = 1, batch_norm = True):
    layers = []
    conv_layer = nn.ConvTranspose2d(in_channels, out_channels, kernel_size,
                           stride = stride, padding = padding, bias = False)
    layers.append(conv_layer)
    
    if batch_norm:
        layers.append(nn.BatchNorm2d(out_channels))
    return nn.Sequential(*layers)
```

```python
#residual blocks
class ResBlock(nn.Module):
    def __init__(self, conv_dim):
        super(ResBlock, self).__init__()
        self.conv1 = conv(in_channels = conv_dim, out_channels = conv_dim, 
                               kernel_size = 3, stride = 1 , padding = 1, batch_norm = True)
        self.conv2 = conv(in_channels = conv_dim, out_channels = conv_dim, 
                               kernel_size = 3, stride = 1 , padding = 1, batch_norm = True)  
    def forward(self, x):
        # y = F(x) + x
        out1 = F.relu(self.conv1(x))
        out2 = x + self.conv2(out1)
        
        
        return out2   
```

```python
class Discriminator(nn.Module):

    def __init__(self, conv_dim=32, n_res_blocks = 3):
        """
        Initialize the Discriminator Module
        :param conv_dim: The depth of the first convolutional layer
        """
        super(Discriminator, self).__init__()
        
        self.conv_dim = conv_dim
        
        #32x32x3 to 16x16x32
        self.conv1 = conv(3, conv_dim, 4, batch_norm = False)
        #16x16x32 to 8x8x64
        self.conv2 = conv(conv_dim, conv_dim*2, 4, batch_norm = True) 
        #8x8x64 to 4x4x128
        self.conv3 = conv(conv_dim*2, conv_dim*4, 4, batch_norm = True)  
        
        #residual blocks
        res_layers = []
        for layer in range(n_res_blocks):
            res_layers.append(ResBlock(conv_dim*4))
        self.resblocks =nn.Sequential(*res_layers)
        
        #average pooling layer
        #self.avgpool = nn.AdaptiveAvgPool2d((1,1))
        self.conv4 = conv(conv_dim*4, conv_dim*8, 4, batch_norm = True)
        
        #classification layer
        self.fc = nn.Linear(conv_dim*8*2*2, 1)

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: Discriminator logits; the output of the neural network
        """
        # define feedforward behavior
        out = F.leaky_relu(self.conv1(x),0.2)
        out = F.leaky_relu(self.conv2(out),0.2)
        out = F.leaky_relu(self.conv3(out),0.2)
        
        #residual blocks
        out = self.resblocks(out)
        
        #global average pooling
        #out = self.avgpool(out)
        out = F.leaky_relu(self.conv4(out))
        
        #flattening and classifaction
        out = out.view(out.size(0), -1)
        out = self.fc(out)
        
        
              
        return out


"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_discriminator(Discriminator)
```

## Generator

The generator should upsample an input and generate a *new* image of the same size as our training data `32x32x3`. This should be mostly transpose convolutional layers with normalization applied to the outputs.

#### Exercise: Complete the Generator class
* The inputs to the generator are vectors of some length `z_size`
* The output should be a image of shape `32x32x3`

```python
class Generator(nn.Module):
    
    def __init__(self, z_size=100, conv_dim=32, n_res_blocks=3):
        """
        Initialize the Generator Module
        :param z_size: The length of the input latent vector, z
        :param conv_dim: The depth of the inputs to the *last* transpose convolutional layer
        """
        super(Generator, self).__init__()
        
        self.conv_dim = conv_dim
        
        self.fc = nn.Linear(z_size, conv_dim*2*2*8)
        self.deconv0 = deconv(conv_dim*8, conv_dim*4,4)      
        #residual blocks
        res_layers = []
        for layer in range(n_res_blocks):
            res_layers.append(ResBlock(conv_dim*4))
        self.resblocks =nn.Sequential(*res_layers)
        
        #decoder of the generator
        #4x4x128 to 8x8x64
        self.deconv1 = deconv(conv_dim*4, conv_dim*2, 4, batch_norm=True)
        #8x8x64 to 16x16x32
        self.deconv2 = deconv(conv_dim*2, conv_dim, 4, batch_norm=True)
        #16x16x32 to 32x32x3
        self.deconv3 = deconv(conv_dim, 3, 4, batch_norm=False)  # no batch_norm on last layer
        

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: A 32x32x3 Tensor image as output
        """
        # define feedforward behavior
        out = self.fc(x)
        out = out.view(-1, self.conv_dim*8, 2, 2) # (Batch_Size, Depth, H, W)
        out = F.relu(self.deconv0(out))
        #residual blocks
        out = self.resblocks(out)
        
        #decoder
        out = F.relu(self.deconv1(out))
        out = F.relu(self.deconv2(out))
        
        # last layer + tanh activation
        out = self.deconv3(out)
        out = F.tanh(out)
        
        
        return out

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_generator(Generator)
```

To help your models converge, you should initialize the weights of the convolutional and linear layers in your model. From reading the [original DCGAN paper](https://arxiv.org/pdf/1511.06434.pdf), they say:

#### Exercise: Complete the weight initialization function

* This should initialize only **convolutional** and **linear** layers
* Initialize the weights to a normal distribution, centered around 0, with a standard deviation of 0.02.
* The bias terms, if they exist, may be left alone or set to 0.

```python
def weights_init_normal(m):
    """
    Applies initial weights to certain layers in a model .
    The weights are taken from a normal distribution 
    with mean = 0, std dev = 0.02.
    :param m: A module or layer in a network    
    """
    # classname will be something like:
    # `Conv`, `BatchNorm2d`, `Linear`, etc.
    
    classname = m.__class__.__name__
    if hasattr(m, 'weight') and (classname.find('Conv') != -1 or classname.find('Linear') != -1):
        nn.init.normal_(m.weight.data, mean=0, std=0.02)
        #nn.init.kaiming_normal_(m.weight.data, a=0, mode='fan_in')
        
        # The bias terms, if they exist, set to 0
        if hasattr(m, 'bias') and m.bias is not None:
            nn.init.constant_(m.bias.data, 0.0)
```

## Discriminator and Generator Losses

Now we need to calculate the losses for both types of adversarial networks.

### Discriminator Losses

> * For the discriminator, the total loss is the sum of the losses for real and fake images, `d_loss = d_real_loss + d_fake_loss`. 
* Remember that we want the discriminator to output 1 for real images and 0 for fake images, so we need to set up the losses to reflect that.


### Generator Loss

The generator loss will look similar only with flipped labels. The generator's goal is to get the discriminator to *think* its generated images are *real*.

#### Exercise: Complete real and fake loss functions

**You may choose to use either cross entropy or a least squares error loss to complete the following `real_loss` and `fake_loss` functions.**

```python
def real_loss(D_out, smooth = False):
    '''Calculates how close discriminator outputs are to being real.
       param, D_out: discriminator logits
       return: real loss'''
    batch_size = D_out.size(0)
    labels = torch.ones(batch_size)
    
    # smooth, real labels = 0.9
    if smooth:
        labels = labels *0.9
        
    # move labels to GPU if available 
    if train_on_gpu:
        labels = labels.cuda()
        
    # using binary cross entropy loss with logits loss
    criterion = nn.BCEWithLogitsLoss()
    
    loss =  criterion(D_out.squeeze() ,labels)
    return loss

def fake_loss(D_out):
    '''Calculates how close discriminator outputs are to being fake.
       param, D_out: discriminator logits
       return: fake loss'''
    batch_size = D_out.size(0)
    labels = torch.zeros(batch_size)
    
    # move labels to GPU if available 
    if train_on_gpu:
        labels = labels.cuda()
        
    # using binary cross entropy loss with logits loss
    criterion = nn.BCEWithLogitsLoss()
    
    loss =  criterion(D_out.squeeze() ,labels)
    return loss
```

## Training

Training will involve alternating between training the discriminator and the generator. You'll use your functions `real_loss` and `fake_loss` to help you calculate the discriminator losses.

* You should train the discriminator by alternating on real and fake images
* Then the generator, which tries to trick the discriminator and should have an opposing loss function


#### Saving Samples

You've been given some code to print out some loss statistics and save some generated "fake" samples.

#### Exercise: Complete the training function

Keep in mind that, if you've moved your models to GPU, you'll also have to move any model inputs to GPU.

```python
def train(D, G, n_epochs, print_every=50):
    '''Trains adversarial networks for some number of epochs
       param, D: the discriminator network
       param, G: the generator network
       param, n_epochs: number of epochs to train for
       param, print_every: when to print and record the models' losses
       return: D and G losses'''
    
    # move models to GPU
    if train_on_gpu:
        D.cuda()
        G.cuda()

    # keep track of loss and generated, "fake" samples, learning rates
    samples = []
    losses = []
    d_scheduler_steps = []
    g_scheduler_steps = []

    # Get some fixed data for sampling. These are images that are held
    # constant throughout training, and allow us to inspect the model's performance
    sample_size=16
    fixed_z = np.random.uniform(-1, 1, size=(sample_size, z_size))
    fixed_z = torch.from_numpy(fixed_z).float()
    # move z to GPU if available
    if train_on_gpu:
        fixed_z = fixed_z.cuda()

    # epoch training loop
    for epoch in range(n_epochs):

        # batch training loop
        for batch_i, (real_images, _) in enumerate(celeba_train_loader):

            batch_size = real_images.size(0)
            real_images = scale(real_images)

            # ===============================================
            #         YOUR CODE HERE: TRAIN THE NETWORKS
            # ===============================================
            
            # 1. Train the discriminator on real and fake images

            d_optimizer.zero_grad()
            
            #train with real images
            if  train_on_gpu:
                real_images = real_images.cuda()
                
            D_real = D(real_images)
            d_real_loss = real_loss(D_real)
            
            #train with fake images
            z = np.random.uniform(-1, 1, size = (batch_size, z_size))
            z = torch.from_numpy(z).float()
            
            if train_on_gpu:
                z = z.cuda()
                
            fake_images = G(z)
            
            D_fake = D(fake_images)
            d_fake_loss = fake_loss(D_fake)
        
            d_loss = d_real_loss + d_fake_loss
            
            d_loss.backward()
            d_optimizer.step()
            
            # 2. Train the generator with an adversarial loss
            
            g_optimizer.zero_grad()
            
            z = np.random.uniform(-1, 1, size = (batch_size, z_size))
            z = torch.from_numpy(z).float()
            
            if train_on_gpu:
                z = z.cuda()
                
            fake_images = G(z)
            D_fake = D(fake_images)
            
            g_loss = real_loss(D_fake)
    
            g_loss.backward()
            g_optimizer.step()
            
            
            # ===============================================
            #              END OF YOUR CODE
            # ===============================================

            # Print some loss stats
            if batch_i % print_every == 0:
                # append discriminator loss and generator loss
                losses.append((d_loss.item(), g_loss.item()))
                # print discriminator and generator loss
                print('Epoch [{:5d}/{:5d}] | d_loss: {:6.4f} | g_loss: {:6.4f}'.format(
                        epoch+1, n_epochs, d_loss.item(), g_loss.item()))
                
        #update learning scheduler
        d_scheduler.step(d_loss.item())
        g_scheduler.step(g_loss.item())
        d_scheduler_steps.append(d_optimizer.param_groups[0]['lr'])
        g_scheduler_steps.append(g_optimizer.param_groups[0]['lr'])


        ## AFTER EACH EPOCH##    
        # this code assumes your generator is named G, feel free to change the name
        # generate and save sample, fake images
        G.eval() # for generating samples
        samples_z = G(fixed_z)
        samples.append(samples_z)
        G.train() # back to training mode

    # Save training generator samples
    with open('train_samples.pkl', 'wb') as f:
        pkl.dump(samples, f)
    
    # finally return losses
    return losses, d_scheduler_steps, g_scheduler_steps
```

Therefore, I think the complicated model will overfit the data. Also, the Discriminator and Generator should be designed properly. At first, I think there were more layers in the Generator (encoders and decoders), so I added more residual blocks on Discriminator, but the d_loss dropped too fast approaching 0 ,and the g_loss became divergent after about 5 epochs. Even I applied some learning rate decay technique, it still happened. Based on the assumption that it's more difficult to train the Generator than the Discriminator, so I decreased the layers in Discriminator and retrained. Although the loss is worse especailly in Discriminator, they are more stable.
3. We all know that it is difficult to reach Nash Equilibrium in GAN, and it was  hard to converge especially for the Generator. I tried many learning rate strategies. Exponential learning rate decay is a quite straightforward method, however, the learning late will be too small for Discriminator. The d_loss will be nearly 0 , but the g_loss will explode. We can't even see the face shape in the output picture. As for the ReduceLROnPlateau way, it's good in other situation, but the g_loss will be more jumpy intrinsically. On the other hand, the Discrinmator is easier to train. d_loss can continuous decrease,so it's learning rate don't need to drop. At the same time learning rate of the Generator will decrease every epoch because g_loss sometimes goes up and is not stable. Finally, I use Cosine Annealing method, it decreased the learning rate, but it will restart after 1 epoch. And, the minimum learning rate is the key, if it is too small the learning will be jumpy and d_loss will be in saddle point or bad local minimum sometimes ,and stop learning. Now, it seems that my loss won't go uncontrollable especially g_loss, and the picture generated can see the face shape clearly, nevertheless, with some acceptable noise
